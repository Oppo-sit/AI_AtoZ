# Get the data

```python
import os
import tarfile
from six.moves import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/" # download base url
HOUSING_PATH = os.path.join("datasets", "housing") # subdirectory has 'housing.tgz'
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):# subdirectory가 없다면,
        os.makedirs(os.path.join(os.getcwd(), housing_path)) # mkdir
    tgz_path = os.path.join(housing_path, "housing.tgz") # save path
    urllib.request.urlretrieve(housing_url, tgz_path) # download 'housing.tgz'
    housing_tgz = tarfile.open(tgz_path) 
    housing_tgz.extractall(path=housing_path) # tgz extraction
    housing_tgz.close()
```

# Split the data

```python
import numpy as np

def split_train_test(data, test_ratio): 
    shuffled_indices = np.random.permutation(len(data)) # index shuffle
    test_set_size = int(len(data) * test_ratio) # set split size
    test_indices = shuffled_indices[:test_set_size] # split test index
    train_indices = shuffled_indices[test_set_size:] # split train index
    return data.iloc[train_indices], data.iloc[test_indices] # return splited data

train_set, test_set = split_train_test(housing, 0.2)
```

```python
display(len(train_set), len(test_set))
```

```python
# solution for prone to new data
from zlib import crc32

def test_set_check(identifier, test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32 # make hash value

def split_train_test_by_id(data, test_ratio, id_column):
    ids= data[id_column] # index = id
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio)) # apply : give hash
    return data.loc[~in_test_set], data.loc[in_test_set] # split data by hash
```

```python
# housing data doesn't have any identifier, so use row index as 'ID'
housing_with_id = housing.reset_index() #adds an 'index' column
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "index")
```

## Custom Transformers

We need to write our own tasks, in case of custom cleanup operations or combining specific attributes
Since Scikit-learn relies on duck taping(not inheritance), 
we need to create a class and implement three methods: fit(), transform(), and fit_transform()
thoes methods could be only return self, 
and we can implement fit_transform easily by scikit-learn class "TransformerMixin" and "BaseEstimator"(no \*\*args/ \*\*kargs)


```python
# Make class for custom transformer combining attributes
from sklearn.base import BaseEstimator, TransformerMixin

#column index
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                        bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
        
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
```

```python
# previous code was hard-coded in finding index, but there are much cleaner way to get thems

col_names = "total_rooms", "total_bedrooms", "population", "households"
rooms_ix, bedrooms_ix, population_ix, households_ix = [
    housing.columns.get_loc(c) for c in col_names
]
```

```python
# Also, 'housing_extra_attribs' is Numpy array, for recovering this to DataFrame, write like this

housing_extra_attribs = pd.DataFrame(
    housing_extra_attribs, 
    columns = list(housing.columns)+["rooms_per_household", "population_per_household"],
    index = housing.index)
housing_extra_attribs.head()
```

## Transformation Pipelines

Before explaining pipeline, we need to know what is 'scaling'<br>
Scaling is regulating scale of all attiributes that would be trained, because of performance drop from scaling difference. (Raw data ordinarily has different scale, and it would be extreme, as the result performance would be dropped.) <br>
There two ways to rescale is 'min-max scailing', 'standardization'. Detailed explanation is left out.
<br><br>
So, now we return to pipeline. Pipeline has the function that helps with such sequences of transformations to execute. In pipeline, implementation method has different meaning. **'fit()'** method execute fit_transform() method until it reaches the final estimator, and **'transform()'** method applies all the transforms to the data in sequence

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler()),
])

housing_num_tr = num_pipeline.fit_transform(housing_num)
```

```python
housing_num_tr
```

```python
# It's better to transform numerical attributes and categorial(text) attributes separately
# Scikit-learn serves 'CoulnmTransformer', which can be handle data by column

from sklearn.compose import ColumnTransformer

num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])

housing_prepared = full_pipeline.fit_transform(housing)

housing_prepared
```
